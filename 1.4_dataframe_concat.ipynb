```python
import os
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
dir_path = "data/for_dataframe_concat"
strincluded = "_NYC_"
output_file_path = dir_path + "/" + f"concat{strincluded}.csv"
```

```python
files_file = [
    f for f in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, f))
]
```

```python
for file in files_file:
    if file.find(strincluded) >= 0 and file != "" and file != None:
        print(file)

        if os.path.isfile(output_file_path) == False:
            df_csv = pd.read_csv(dir_path + "/" + file, header=0)
            df_csv.to_csv(output_file_path, index=False)
            del df_csv
        else:
            df_csv = pd.read_csv(output_file_path, header=0)
            df_csv_concat = pd.read_csv(dir_path + "/" + file, header=0)
            df_csv_concat.dropna(subset="Preferred Skills", inplace=True) # drop rows with NaN in "Preferred Skills" column
            print(df_csv_concat.shape)

            df_output_csv = pd.concat([df_csv, df_csv_concat], sort=False)
            df_output_csv.drop_duplicates(subset=['Job ID', 'Agency'], keep="last", inplace=True) # drop duplicates
            df_output_csv.sort_values(['Job ID', 'Agency'], ascending=[True, True], inplace=True) # sort by "Job ID" and "Agency"
            df_output_csv.reset_index(drop=True, inplace=True) # reset index
            df_output_csv.to_csv(output_file_path, index=False)

            del df_csv
            del df_csv_concat
            del df_output_csv
```
