# rename column name

```python
df_response = pd.read_csv('./data/df_response.csv', header=0)
df_response = df_response.rename(columns={'jobid': 'Job ID'})
df_response.loc[0:, ["Job ID", "requirement", "reason"]]
```

```python
df_jobs = pd.read_csv('./data/Jobs_NYC_Postings.csv', header=0)
df_jobs = df_jobs.loc[0:,["Job ID", "Agency", "Business Title", "Job Category"]]
df_jobs
```

```python
df_jobs[df_jobs['Job ID'].isin([606346,571361])]
```

# left outer join

```python
df_jobs_and_reason = pd.merge(df_jobs, df_response, on="Job ID", how="left")
df_jobs_and_reason
```

```python
df_jobs_and_reason[df_jobs_and_reason['reason'].isna()]
```

```python
df_jobs_and_reason
```

# delete rows if specified column's value is NaN

```python
df_jobs_and_reason.dropna(subset=['reason'], inplace=True)
df_jobs_and_reason
```

# update specified column's value on given condition

```python
for index, row in df_jobs_and_reason[0:].iterrows():
    if row["Job ID"] == 571361:
        new_value = "Architect / " + row["Business Title"]
        df_jobs_and_reason.at[index, 'Business Title'] = new_value
df_jobs_and_reason
```

# Select rows where the 'column_name' values are in a list of values

```python
# Select rows where the 'column_name' values are in a list of values
df_jobs_and_reason = df_jobs_and_reason[df_jobs_and_reason['Business Title'].isin(['Architect / Executive Program Manager'])]
df_jobs_and_reason
```

# reset index

```python
df_jobs_and_reason.reset_index(drop=True, inplace=True)
df_jobs_and_reason
```
