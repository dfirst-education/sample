```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("./images/aws.png")
```

```python
fig, ax = plt.subplots()
fig.set_facecolor((0.8, 0.4, 0.4, 0.5))
ax.set_facecolor((0.8, 0.4, 0.4, 0.5))
ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # OpenCV sorts colors in RGB order since they are in GBR order
plt.show()
```

```python
mask = np.all(img[:,:,:] == [255, 255, 255], axis=-1)
dst = cv2.cvtColor(img, cv2.COLOR_BGR2BGRA)
dst[mask,3] = 0
cv2.imwrite("./images/aws_transparent.png", dst)
```

```python
img2 = cv2.imread("./images/aws_transparent.png")

fig, ax2 = plt.subplots()
fig.set_facecolor((0.8, 0.4, 0.4, 0.5))
ax2.set_facecolor((0.7, 0.9, 0.8, 0.5))
ax2.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)) # OpenCV sorts colors in RGB order since they are in GBR order
ax2.plot()
plt.show()
```
