```python
from sklearn.linear_model import LinearRegression # import LinearRegression
from sklearn.datasets import fetch_california_housing # import fetch_california_housing dataset
from sklearn.model_selection import train_test_split # import train_test_split to split the dataset
import pickle

# data = fetch_california_housing() # load the dataset
# with open(f'../data/fetch_california_housing.pkl', 'wb') as f:
# 	pickle.dump(data, f)
with open(f'../data/fetch_california_housing.pkl', 'rb') as f:
	data = pickle.load(f)

train_X, test_X, train_y, test_y = train_test_split(
    data.data, data.target, random_state=42) # split the dataset into training and testing datasets

model = LinearRegression() # create a LinearRegression model
model.fit(train_X, train_y) # train the model using the training dataset

pred_test_y = model.predict(test_X) # predict the target values for the test dataset
print(pred_test_y)

model.score(test_X, test_y)
```
