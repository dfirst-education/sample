```python
def test_func():
    print("Hello, World!")
    raise Exception("This is a test exception")
```

```python
import time # for sleep
import traceback # for error message
for _ in range(10): # max retry count
    try:
        test_func()
    except Exception as e:
        print("An error occurred", _)
        print(traceback.format_exc()) # error message
        time.sleep(30) # wait 30 seconds
    else:
        print("Succeeded! The process was completed without any errors. Loop was ended.")
        break
else:
    print("max retry count exceeded. The process was not completed.")
    raise Exception ("max retry count exceeded. The process was not completed.")
```
